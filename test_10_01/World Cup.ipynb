```python
Teams=['India','Pakistan','Australia','Newzealand','Ireland','England','West Indies','Bangladesh','Sri Lanka','Zimbabwe','South Africa','Afghanistan','UAE','China','Japan','Brazil']



SectionA=['India','Pakistan','Australia','Newzealand']
SectionB=['Ireland','England','West Indies','Bangladesh']
SectionC=['Sri Lanka','Zimbabwe','South Africa','Afghanistan']
SectionD=['UAE','China','Japan','Brazil']


def whoWon(x):
    '''returns a list of winners out of 4 blocks'''
    won={}
    winners=[]
    
    
    #Arranging matches and populating won dictionary
    for i in range(3):
        for j in range(i+1,4):
            print('who won between',x[i],'and',x[j])
            winner=input()
            if winner=='draw':
                won[x[i]]=won[x[i]]+1
                won[x[j]]=won[x[j]]+1
            elif winner==x[i]:
                if x[i] in won:
                    won[x[i]]=won[x[i]]+2
                else:
                    won[x[i]]=2
            elif winner==x[j]:
                if x[j] in won:
                    won[x[j]]=won[x[j]]+2
                else:
                    won[x[j]]=2
    count=0
    for k in sorted(won, key=won.get, reverse=True):     #Sorting the dictionary
           if count<2: 
                winners.append(k)
                count=count+1
    return winners

#Round 1
Winseca=whoWon(SectionA)
Winsecb=whoWon(SectionB)
Winsecc=whoWon(SectionC)
Winsecd=whoWon(SectionD)

#Round 2 
e=Winseca+Winsecb
f=Winsecc+Winsecd
Winnerine=whoWon(e)
Winnerinf=whoWon(f)

#Final Battle
Champion=Winnerine+Winnerinf
final=whoWon(Champion)

#And it is the time to declare the winner! 
print("Final winner is",final[0])

```
